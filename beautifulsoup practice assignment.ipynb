```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

Solution 1

```python
page1=requests.get('https://www.wikipedia.org/')
page1
```

```python
soup1=BeautifulSoup(page1.content)
soup1
```

```python
header=[]
for i in soup1.find_all(['h1','h2','h3','h4','h5','h6']):
    header.append(i.text)
    
header
```

Solution 2

```python
page2=requests.get('https://www.imdb.com/search/title/?sort=user_rating,desc&groups=top_100')
page2
```

```python
soup2=BeautifulSoup(page2.content)
soup2
```

```python
name=[]

for i in soup2.findall('h3',class_='ipc-title__text'):
    name.append(i.text)
    
name
```

solution 3

```python
page3=requests.get('https://www.dineout.co.in/ahmedabad-restaurants?loc=Raipur,%20East%20Ahmedabad')
page3
```

```python
soup3=BeautifulSoup(page3.content)
soup3
```

```python
name = []

for i in soup3.find_all('div',class_='restnt-info cursor'):
    name.append(i.text)
    
name  
```

```python
cuisine = []

for i in soup3.find_all('span',class_='double-line-ellipsis'):
    cuisine.append(i.text)
    
cuisine
```

```python
location = []

for i in soup3.find_all('div',class_='restnt-loc ellipsis'):
    location.append(i.text)
    
location
```

```python
rating = []

for i in soup3.find_all('div',class_='restnt-rating rating-4'):
    rating.append(i.text)
for i in soup3.find_all('div',class_='restnt-rating rating-5'):
    rating.append(i.text)
    
rating
```

```python
images = []

for i in soup3.find_all('img',class_='no-img'):
    images.append(i['data-src'])
    
images
```

```python
print(len(images),len(rating),len(location),len(cuisine),len(name))
```

```python
df=pd.DataFrame({'Name':name,'Location':location,'Cuisine':cuisine,'Image url':images})
df
```
