```python

product_name=[]
product_url=[]
```

```python
start=0
end=3
for page in range(start,end):
    producttag=driver.find_elements(By.XPATH,'//div[@class="a-section a-spacing-small puis-padding-left-small puis-padding-right-small"]/div/h2/a/span')
    for i in producttag:
        product_name.append(i.text)
    next_button=driver.find_elements(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div[1]/div/span[1]/div[1]/div[69]/div/div/span/a[3]')
```

```python
start=0
end=3
for page in range(start,end):
    urltag=driver.find_elements(By.XPATH,'//div[@class="a-section a-spacing-none a-spacing-top-small s-title-instructions-style"]/h2/a')
    for i in urltag:
        product_url.append(i.get_attribute('href'))
    next_button=driver.find_elements(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div[1]/div/span[1]/div[1]/div[69]/div/div/span/a[3]')
```

```python
brand_name=[]
exchange=[]
price=[]
delivery=[]
availability=[]
for i in product_url:
    driver.get(i)
    time.sleep(5)
    
    try:
        availtag=driver.find_element(By.XPATH,'/html/body/div[2]/div/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[4]/div/div[5]/div/div[1]/span')
        availability.append(availtag.text)
    except NoSuchElementException:
        availability.append('-')
        
        
    try:
        delivertag=driver.find_element(By.XPATH,'/html/body/div[2]/div/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[4]/div/div[3]/div[10]/div[1]/div/div/div/span/span[1]')
        delivery.append(delivertag.text)
    except NoSuchElementException:
        delivery.append('-')
        
        
    
    try:
        pricetag=driver.find_element(By.XPATH,'/html/body/div[2]/div/div[5]/div[3]/div[4]/div[13]/div/div/div[4]/div[1]/span[3]/span[2]/span[2]')
        price.append(pricetag.text)
    except NoSuchElementException:
        price.append('-')
        
        
    
    try:
        brandtag=driver.find_element(By.XPATH,'/html/body/div[2]/div/div[5]/div[3]/div[4]/div[48]/div/table/tbody/tr[1]/td[2]/span')
        brand_name.append(brandtag.text)
    except NoSuchElementException:
        brand_name.append('-')
        
        
        
        
    try:
        exchtag=driver.find_element(By.XPATH,'/html/body/div[2]/div/div[5]/div[3]/div[4]/div[24]/div[2]/div/div/div/div[2]/div/ol/li[3]/div/span/div[2]/span')
        exchange.append(exchtag.text)
    except NoSuchElementException:
        exchange.append('-')
        
        
```

```python
len(availability),len(brand_name),len(price),len(exchange),len(delivery),len(product_name),len(product_url)
```

```python
df=pd.DataFrame({"Product Name":product_name,"Brand":brand_name,'Price':price,'Avialability':availability,'Delivery':delivery,'Exchange':exchange,'Product Url':product_url})
```
