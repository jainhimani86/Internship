```python
driver=webdriver.Chrome()
```

```python
driver.get('https://github.com/')
```

```python
source=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
source.click()
```

```python
trending=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a')
trending.click()
```

```python
title=[]
description=[]
count=[]
language=[]
```

```python
titletag=driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a')
for i in titletag:
    title.append(i.text)
```

```python
title
```

```python
desctag=driver.find_elements(By.XPATH,'//p[@class="col-9 color-fg-muted my-1 pr-4"]')
for i in desctag:
    description.append(i.text)
```

```python
description
```

```python
url=[]
urltag=driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a')
for i in urltag:
    url.append(i.get_attribute('href'))
```

```python
for i in url:
    driver.get(i)
    time.sleep(5)

    try:
        langtag=driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/main/turbo-frame/div/div/div/div[2]/div[2]/div/div[7]/div/ul/li[1]/a/span[1]')
        language.append(langtag.text)
    except NoSuchElementException:
        language.append("-")
        
    
    try:
        counttag=driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/main/turbo-frame/div/div/div/div[2]/div[2]/div/div[6]/div/h2/a/span')
        count.append(counttag.text)
    except NoSuchElementException:
        count.append('-')
```

```python
df=pd.DataFrame({'Repository Title':title,'Repository Description':description,'Contributors count':count,'Language used':language})
```

```python
df
```
